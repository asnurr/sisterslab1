```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('C:\sisterslab1\sisterslab1\data\starbucks.csv')
df.head()
```

```python
# Veri setinin genel istatistiklerini görüntüle
print(df.describe().T)
```

```python
print(df.isnull().sum())
```

```python
# Sayısal olmayan sütunları çıkar
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
numeric_data = df[numeric_columns]

# Korelasyon matrisini oluştur
correlation_matrix = numeric_data.corr()

# Korelasyon matrisini görselleştir
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Sayısal Değişkenler Arası Korelasyon')
plt.show()
```

```python
# Kategorik değişkenlerin dağılımını görselleştirin
plt.figure(figsize=(12, 6))
sns.countplot(x='Beverage_category', data=df, palette='viridis')
plt.title('Starbucks Ürün Kategorileri Dağılımı')
plt.xticks(rotation=45)
plt.show()
```
